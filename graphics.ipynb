```python
import pandas as pd
```

```python
conferences = pd.DataFrame({
    "Conferência": ["RecSys", "SIGIR", "CIKM", "WWW", "TOIS", "KDD", "WSDM", "IJCAI", "Informations Science",
                    "Neural Networks", "AAAI", "Neurocomputing", "IEEE. Data Engineering", "IEEE. Big Data",
                    "IEEE Access", "IEEE. Social Systems", "IEEE. Learning Systems", "ACL", "IJCNLP"],
    "Contagem": [6, 6, 5, 5, 4, 4, 4, 4, 3, 2, 2, 3, 2, 1, 1, 1, 1, 1, 1]
})

conferences = conferences.sort_values(by="Contagem", ascending=False)
```

```python
conferences['Contagem'].sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# plt.style.use("grayscale")
```

```python


ax = sns.barplot(x=conferences['Conferência'], y=conferences['Contagem'], color='gray')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)

sns.set_style("white")

plt.ylabel("# de artigos")
plt.show()
```

```python
fit, ax = plt.subplots()

ax.set_facecolor('white')

ax.bar(conferences['Conferência'], conferences['Contagem'], color='gray')

# bars = plt.bar(conferences['Conferência'], conferences['Contagem'])
plt.xticks(rotation=90)
plt.xlabel("Conferências"); plt.ylabel("# de artigos")
plt.show()  
```
